```python
import pandas as pd
import numpy as np
import os
```

```python
#locate the directory where the excel files are
excel_dir = r'C:\Users\Dell\Documents\july recon\ATS SPECIAL'
excel_dir1 = r'C:\Users\Dell\Documents\july recon\INWARDS JULY 2018\INWARDS JULY 2018'
excel_dir2 = r'C:\Users\Dell\Documents\july recon\OUTWARDS JULY 2018\OUTWARDS JULY 2018'
excel_dir3 = r'C:\Users\Dell\Documents\july recon\ATS ITEMS'
```

```python
#use the os module to list all files inn the directory
all_files = os.listdir(excel_dir)
all_files1 = os.listdir(excel_dir1)
all_files2 = os.listdir(excel_dir2)
all_files3 = os.listdir(excel_dir3)
```

```python
#create an empty list to store the excel file names
excel_files = []
excel_files1 = []
excel_files2 = []
excel_files3 = []
```

```python
#loop through all the files and keep only those that end with '.xlsx'
for file in all_files:
    if file.endswith('.xlsx'):
        excel_files.append(file)

#loop for inward
for file in all_files1:
    if file.endswith('.xlsx'):
        excel_files1.append(file)
        
#loop for outward
for file in all_files2:
    if file.endswith('.xlsx'):
        excel_files2.append(file)
        
#loop for ATS
for file in all_files3:
    if file.endswith('.xlsx'):
        excel_files3.append(file)
```

```python
#initialize an empty list to store dataframes
dfs = []
dfs1 = []
dfs2 = []
dfs3 = []
```

```python
#loop through each excel file and read it into a dataframe
for file in excel_files:
    file_path = os.path.join(excel_dir, file)
    df = pd.read_excel(file_path)
    dfs.append(df)
```

```python
#loop for inward
for file in excel_files1:
    file_path1 = os.path.join(excel_dir1, file)
    df1 = pd.read_excel(file_path1)
    dfs1.append(df1) 
```

```python
#loop for outward
for file in excel_files2:
    file_path2 = os.path.join(excel_dir2, file)
    df2 = pd.read_excel(file_path2)
    dfs2.append(df2)
```

```python
#loop for ats
for file in excel_files3:
    file_path3 = os.path.join(excel_dir3, file)
    df3 = pd.read_excel(file_path3)
    dfs3.append(df3)
```

```python
#concatenate the list of dataframes into one (spcial clearance)
merged_df = pd.concat(dfs, axis = 0, ignore_index = True)

#concatenate for inward
merged_df1 = pd.concat(dfs1, axis = 0, ignore_index = True)

#concatenate for outward
merged_df2 = pd.concat(dfs2, axis = 0, ignore_index = True)

#concate for ATS
merged_df3 = pd.concat(dfs3, axis = 0, ignore_index = True)
```

```python
# Use drop to delete the empty column
merged_df1.drop('Unnamed: 8', axis=1, inplace=True)
```

```python
# Define a list of new column names to be assigned to the dataframe
new_column_names1 = ['Date', 'Debit A/C Number', 'Debit A/C Name', 'Credit A/C Number', 'Credit A/C Name','BIC', 'Batch Reference from ATS', 'Amount', 'Serial Number']

# Assign the new column names to the DataFrame
merged_df1.columns = new_column_names1

# Define a list of new column names to be assigned to the dataframe
new_column_names2 = ['Date', 'Debit A/C Number', 'Debit A/C Name', 'Credit A/C Number', 'Credit A/C Name','BIC', 'Item Reference from ATS', 'Amount', 'Serial Number']

# Assign the new column names to the DataFrame
merged_df2.columns = new_column_names2
```

```python
# convert the inward amount column into negative
merged_df1['Amount'] = merged_df1['Amount'] * -1
```

```python
# Because the header of every file was also placed as a data field in the dataframe we are removing it here
#inward
merged_df1 = merged_df1[~merged_df1['Date'].str.contains('Date', case=False, na=False)]

# Reset the index after removing the rows
merged_df1.reset_index(drop=True, inplace=True)

# outward
merged_df2 = merged_df2[~merged_df2['Date'].str.contains('Date', case=False, na=False)]

# Reset the index after removing the rows
merged_df2.reset_index(drop=True, inplace=True)
```

```python
# Filter rows where 'Amount' column is null (missing value)
rows_with_missing_amount = merged_df1[merged_df1['Serial Number'].isnull()]

# Save the filtered rows to a new DataFrame and reset index
filtered_rows_df = rows_with_missing_amount.reset_index(drop=True)

# Define the output Excel file path for rows with missing amount
output_excel_path = 'rows_with_missing_amount.xlsx'

# Save the filtered rows to an Excel file
filtered_rows_df.to_excel(output_excel_path, index=False)

print(f"Rows with missing amount saved to '{output_excel_path}'")
```

```python
# Convert the 'Amount' column to float for positive amounts dataframe
merged_df101['Amount'] = merged_df101['Amount'].astype(float)

# Convert the 'Amount' column to float for negative amounts dataframe
#negative_amount_df['Amount'] = negative_amount_df['Amount'].astype(float)

# Convert the 'Amount' column to float for merged_df2 (outward) dataframe
merged_df2['Amount'] = merged_df2['Amount'].astype(float)

# Convert the 'Amount' column to float for merged_df101 (inward) dataframe
#merged_df101['Amount'] = merged_df101['Amount'].astype(float)
```

```python
#concatenate transfer reference with amount to create a primary key for the the special clearance data
merged_df['CONCATENATE'] = merged_df.apply(lambda row:str(row['TRANSACTION REFERENCE']) + '' + str(row['AMOUNT']), axis = 1)

#concatenate for inward
merged_df101['CONCATENATE'] = merged_df101.apply(lambda row:str(row['Serial Number']) + '' + str(row['Amount']), axis = 1)


#concatenate for outward
merged_df2['CONCATENATE'] = merged_df2.apply(lambda row:str(row['Serial Number']) + '' + str(row['Amount']), axis = 1)
```
